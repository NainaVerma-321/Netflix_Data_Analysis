```python
#now we need to filter out data because all the data is of no use
#eliminate overview ,language ,poster_url

#DATA PREPROCESSING
```

```python
#we need to convert the datatype of release date from object to date and time

df['Release_Date']=pd.to_datetime(df['Release_Date'])

print(df['Release_Date'].dtypes)
```

```python
df['Release_Date']=df['Release_Date'].dt.year

df['Release_Date'].dtypes
```

```python
df.head()
```

```python
#dropping columns
df.drop(['Overview','Original_Language','Poster_Url'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#we are categorizing the Vote_Average column

def categorize_col(df,col,labels):
  edges= [df[col].describe()['min'],
          df[col].describe()['25%'],
          df[col].describe()['50%'],
          df[col].describe()['75%'],
          df[col].describe()['max']]

  df[col] =pd.cut(df[col],edges,labels=labels,duplicates='drop')
  return df
```

```python
labels=['not_popular','below_average','average','popular']

categorize_col(df,'Vote_Average',labels)

df['Vote_Average'].unique()
```

```python
df.head()
```

```python
#wanna see movies with different tags

df['Vote_Average'].value_counts()
```

```python
#the movies that are not popular are of no use to show
df.dropna(inplace=True)

df.isna().sum()
```

```python
df.head()
```

```python
#we need to remove whitespace and write the genre in separate line s
df['Genre']=df['Genre'].str.split(',')
df=df.explode('Genre').reset_index(drop=True)
df.head()
```

```python
#casting column into category

df['Genre']=df['Genre'].astype('category')
df['Genre'].dtypes
```
